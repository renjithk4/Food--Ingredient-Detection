```python
import json

# Load the JSON file into a Python dictionary
with open('/content/drive/MyDrive/1_food/All Text data/train_data.json', 'r') as file:
    data1 = json.load(file)
```

```python
# Initialize a list to store the processed data
processed_data = []

# Iterate through each item in the list
for item in data1:
    recipe_id = item[0]
    details = item[1]

    # Create a dictionary for each recipe
    recipe_dict = {'recipe_id': recipe_id}

    # Add details to the dictionary
    for key, value in details.items():
        if isinstance(value, list):
            # Handle list types by joining them with a delimiter or keeping them as lists
            recipe_dict[key] = ', '.join(map(str, value)) if value else None
        else:
            recipe_dict[key] = value

    # Append the processed dictionary to the list
    processed_data.append(recipe_dict)

# Convert the list of dictionaries into a DataFrame
df1 = pd.DataFrame(processed_data)

# Print the DataFrame
print(df1.head())
```

```python
import pandas as pd
import json
```

```python
# Normalize ingredients and instructions
ingredients_df = pd.json_normalize(data, 'ingredients', ['id', 'title', 'url', 'partition'], record_prefix='ingredient_')
instructions_df = pd.json_normalize(data, 'instructions', ['id', 'title', 'url', 'partition'], record_prefix='instruction_')

# Combine the dataframes if needed, otherwise they can be kept separate
combined_df = pd.concat([ingredients_df, instructions_df], axis=0, ignore_index=True)

# Show the dataframe
print(combined_df)
```

```python
import pandas as pd

# Assuming the DataFrame 'combined_df' has already been created from the JSON data

# Step 1: Extract the 'title' column into a list
title_list = combined_df['title'].tolist()

# Step 2: Get the unique elements
unique_titles = list(set(title_list))

# Step 3: Get the count of unique titles
unique_count = len(unique_titles)
print(f"Number of unique titles: {unique_count}")

# Step 4: Save the unique titles into a CSV file
unique_titles_df = pd.DataFrame(unique_titles, columns=['Unique Titles'])
unique_titles_df.to_csv('unique_titles.csv', index=False)

# If you are in Google Colab, you can download the file using:
from google.colab import files
files.download('unique_titles.csv')
```

```python
import pandas as pd

# Assuming the DataFrame 'combined_df' is already created and has the following columns:
# 'ingredient_text', 'id', 'title', 'url', 'partition', 'instruction_text'

# Step 1: Select the 'title' and 'ingredient_text' columns
pairs_df = combined_df[['title', 'ingredient_text']]

# Step 2: Drop duplicates to get unique pairs
unique_pairs_df = pairs_df.drop_duplicates()

# Step 3: Display the unique pairs
print(unique_pairs_df)

# Step 4: Save the unique pairs to a CSV file
unique_pairs_df.to_csv('unique_title_ingredient_pairs.csv', index=False)
```

```python
# Drop rows where either 'ingredient_text' or 'title' have NaN values
cleaned_pairs_df = unique_pairs_df.dropna(subset=['ingredient_text', 'title'])

# Display the cleaned DataFrame
print(cleaned_pairs_df)
```
